```python
from qiskit import QuantumCircuit
from qiskit.quantum_info import Statevector
from qiskit.visualization import plot_state_qsphere
```

```python
sv = Statevector.from_label('10') # 11 for -ve, while 00 and 01 for 00 & 11 states
plot_state_qsphere(sv.data)
```

```python
mycircuit = QuantumCircuit(2)
mycircuit.cx(0,1)
mycircuit.draw('mpl')
```

```python
new_sv = sv.evolve(mycircuit)
print(new_sv)
plot_state_qsphere(new_sv.data)
```

```python
counts1 = new_sv.sample_counts(shots=1)
counts2 = new_sv.sample_counts(shots=10)
counts3 = new_sv.sample_counts(shots=100)
counts4 = new_sv.sample_counts(shots=1000)

from qiskit.visualization import plot_histogram
```

```python
plot_histogram(counts1)
```
